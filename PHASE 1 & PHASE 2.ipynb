```python
!pip install beautifulsoup4 requests

import requests
from bs4 import BeautifulSoup
import csv

# Send a GET request to the IMDb URL
url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
response = requests.get(url)

# Create BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Extract movie details from the parsed HTML
movies = soup.find_all('div', class_='lister-item mode-advanced')

# Define the CSV file names
first_csv_file = 'movies_data.csv'

# Define the field names for the first CSV file
fieldnames = ['Sno', 'Movie Name', 'Director1', 'Director2', 'Duration', 'Year', 'Ratings', 'Metascore']

# Create and open the first CSV file in write mode
with open(first_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    # Write the header row
    writer.writeheader()

    # Iterate over each movie
    for i, movie in enumerate(movies, start=1):
        # Extract movie details
        title = movie.h3.a.text
        directors = movie.find_all('p')[2].find_all('a')
        director1 = directors[0].text if len(directors) > 0 else ""
        director2 = directors[1].text if len(directors) > 1 else ""
        duration = movie.find('span', class_='runtime').text.strip(' min')
        year = movie.find('span', class_='lister-item-year').text.strip('()')
        ratings = movie.find('div', class_='inline-block ratings-imdb-rating').strong.text
        metascore = movie.find('span', class_='metascore').text.strip() if movie.find('span', class_='metascore') else ""

        # Write movie data to the CSV file
        writer.writerow({
            'Sno': i,
            'Movie Name': title,
            'Director1': director1,
            'Director2': director2,
            'Duration': duration,
            'Year': year,
            'Ratings': ratings,
            'Metascore': metascore
        })

print("First CSV file created successfully.")

```

```python
second_csv_file = 'movies_data_2.csv'
# Define the field names for the second CSV file
fieldnames = ['Movie Name', 'Star1', 'Star2', 'Star3', 'Star4', 'Votes', 'Genre1', 'Genre2', 'Genre3', 'Gross Collection', 'Popularity', 'Certification']

# Create and open the second CSV file in write mode
with open(second_csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    # Write the header row
    writer.writeheader()

    # Iterate over each movie
    for movie in movies:
        # Extract movie details
        title = movie.h3.a.text
        stars = movie.find('p', class_='').find_all('a')[0:4]
        star1 = stars[0].text if len(stars) > 0 else ""
        star2 = stars[1].text if len(stars) > 1 else ""
        star3 = stars[2].text if len(stars) > 2 else ""
        star4 = stars[3].text if len(stars) > 3 else ""
        votes = movie.find('span', attrs={'name': 'nv'})['data-value']
        genres = movie.find('span', class_='genre').text.strip().split(', ')
        genre1 = genres[0] if len(genres) > 0 else ""
        genre2 = genres[1] if len(genres) > 1 else ""
        genre3 = genres[2] if len(genres) > 2 else ""
        gross_collection = movie.find('span', attrs={'name': 'nv'})['data-value'] if movie.find('span', attrs={'name': 'nv'}) else ""
        popularity = movie.find('span', class_='text-muted').find_all('span')[-1].text.strip() if movie.find('span', class_='text-muted').find_all('span') else ""
        certification = movie.find('span', class_='certificate').text.strip() if movie.find('span', class_='certificate') else ""

        # Write movie data to the CSV file
        writer.writerow({
            'Movie Name': title,
            'Star1': star1,
            'Star2': star2,
            'Star3': star3,
            'Star4': star4,
            'Votes': votes,
            'Genre1': genre1,
            'Genre2': genre2,
            'Genre3': genre3,
            'Gross Collection': gross_collection,
            'Popularity': popularity,
            'Certification': certification
        })

print("Second CSV file created successfully.")
```

```python
import sqlite3

# Create SQLite database and establish connection
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()

# Subqueries in Python

# 1) Display all details from both tables where movie names are the same.
subquery1 = '''SELECT *
               FROM Table1
               INNER JOIN Table2 ON Table1."Movie Name" = Table2."Movie Name"'''
cursor.execute(subquery1)
result1 = cursor.fetchall()
print("Subquery 1 Result:")
for row in result1:
    print(row)

# 2) Display all movie names, Director, ratings, and gross collection where the genre is action.
subquery2 = '''SELECT Table1."Movie Name", Table1."Director1", Table1.Ratings, Table2."Gross Collection"
               FROM Table1
               INNER JOIN Table2 ON Table1."Movie Name" = Table2."Movie Name"
               WHERE Table2.Genre1 = 'Action' OR Table2.Genre2 = 'Action' OR Table2.Genre3 = 'Action' '''
cursor.execute(subquery2)
result2 = cursor.fetchall()
print("Subquery 2 Result:")
for row in result2:
    print(row)

# 3) Display all details from both tables with the highest gross collection.
subquery3 = '''SELECT *
               FROM Table1
               INNER JOIN Table2 ON Table1."Movie Name" = Table2."Movie Name"
               WHERE Table2."Gross Collection" = (SELECT MAX("Gross Collection") FROM Table2)'''
cursor.execute(subquery3)
result3 = cursor.fetchall()
print("Subquery 3 Result:")
for row in result3:
    print(row)

# 4) Display all details from both tables with the highest ratings.
subquery4 = '''SELECT *
               FROM Table1
               INNER JOIN Table2 ON Table1."Movie Name" = Table2."Movie Name"
               WHERE Table1.Ratings = (SELECT MAX(Ratings) FROM Table1)'''
cursor.execute(subquery4)
result4 = cursor.fetchall()
print("Subquery 4 Result:")
for row in result4:
    print(row)

# 5) Display all details from both tables with the lowest gross collection and lowest ratings.
subquery5 = '''SELECT *
               FROM Table1
               INNER JOIN Table2 ON Table1."Movie Name" = Table2."Movie Name"
               WHERE Table2."Gross Collection" = (SELECT MIN("Gross Collection") FROM Table2)
                   AND Table1.Ratings = (SELECT MIN(Ratings) FROM Table1)'''
cursor.execute(subquery5)
result5 = cursor.fetchall()
print("Subquery 5 Result:")
for row in result5:
    print(row)

# Close the connection to the SQLite database
conn.close()
```
